```python
# merge the vac_2020 dataframe with the vac_2021 dataframe on the common column "Corridor"

vac_2020_2021 = pd.merge(
    vac_2020, vac_2021, on="Corridor")

# remove any columns with NaN ("Not a Number") - used for missing values, by using .dropna()
vac_2020_2021 = vac_2020_2021.dropna()
vac_2020_2021
```

```python
# Determine which measure of central tendency is most appropriate to describe the Population
plt.hist(vac_2020_2021['2020 Business vacancies <24 months'])
plt.xlabel('Vacancies')
plt.ylabel('Counts')
plt.show()
print(vac_2020_2021['2020 Business vacancies <24 months'].mean())
print(vac_2020_2021['2020 Business vacancies <24 months'].median())
print(vac_2020_2021['2020 Business vacancies <24 months'].mode())
vac_mean = (vac_2020_2021['2020 Business vacancies <24 months'].mean())
print(f"The mean business vacancy <24 months in 2020 is {vac_mean}")
```
